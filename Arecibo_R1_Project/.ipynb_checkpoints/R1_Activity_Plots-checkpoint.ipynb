# Functions

```python
#Kenzie Nimmo's ChiSq code:

def chisq(N,exp):
    """
    N = binned number of bursts
    exp = exposure time per bin
    """
    p=float(np.sum(N)/np.sum(exp))

    E = np.zeros_like(N)
    for binno in range(len(N)):
        E[binno] = p*exp[binno]
    
    chisq_stat=np.sum((N-E)**2/(E))
    
    return chisq_stat


def folding(beginp,endp,bins,burst_times,begin_times,end_times):
    """
    beginp and endp are the beginning and end trial periods to fold in days
    bins is an integer number of bins to divide your period into
    burst_times is an array of the observation days which contains a burst
    begin_times is an array of the observation begin times with AND without bursts
    end_times is an array of the observation end times with AND without bursts
    """
    freqs=np.arange(1./float(endp),1./float(beginp),(1./float(bins))/(np.max(burst_times)-np.min(burst_times)))
    #print(freqs)
    periods = np.sort(1./freqs) #days
    #print(periods)
    
    #fig=plt.figure(figsize=(4,3))
    #ax=fig.add_subplot(111)
    #ax.plot(periods,marker='o',label='periods')
    #ax.plot(freqs,marker='o',label='frequencies')
    #ax.legend()
    #plt.show()
    
    t0=begin_times.min()
    tend=end_times.max()

    chisq_values=[]
    ps=[]
    #for pind,p in enumerate(tqdm(periods)): #removed tqdm for now
    for pind,p in enumerate(periods):
        bin_length= p/float(bins) #days
        binned=np.zeros(bins)
        exposure=np.zeros(bins)
        for i in range(int(np.ceil((tend-t0)/bin_length))): #chop up into bins
            if i==0:
                begin_time = t0 + i*bin_length
            else: begin_time = end_time
            end_time = begin_time + bin_length
            above=np.where(burst_times >= begin_time) 
            below=np.where(burst_times < end_time )
            count=len(np.intersect1d(above,below)) #find the bursts within the begin_time and end_time of the bin
            binned[(i)%bins]+=count #adding a count each time a burst falls in a particular bin
            # compute the exposure
            time_before=np.where(begin_times < end_time)
            time_after=np.where(end_times > begin_time)
            overlap_with_bin=np.intersect1d(time_before,time_after)
            for over in range(len(overlap_with_bin)):
                if begin_times[overlap_with_bin[over]] <= begin_time:
                    begin_exp = begin_time
                else: begin_exp = begin_times[overlap_with_bin[over]]
                if end_times[overlap_with_bin[over]] >= end_time:
                    end_exp = end_time
                else: end_exp = end_times[overlap_with_bin[over]]
            exp = ((end_exp-begin_exp)*24.*3600.) 
            exposure[(i)%bins]+=exp #adding exposure per bin in seconds 
        #print(chisq(binned,exposure))
        chisqp=float(chisq(binned,exposure))/(bins-1)
        chisq_values.append(chisqp)
        ps.append(p)
    return chisq_values, ps

def TallyBursts(ObservingTimes,ObservingDurations,BurstMJDs):
    """
    For papers which provide burst data but not observing information, we must find the observing 
    files where these bursts were stored on dop263.
    
    The bursts must then be tallied into their respective observations. This function does this.
    
    INPUTS:
    
    ObservingTimes     : [array of astropy mjd values] List of observation start times which
                         contained bursts. Must be calculated from observing data.
    ObservingDurations : [array of astropy seconds values] List of observation durations in u.s.
                         Must be calculated from observing data.
    BurstMJDs          : [array of floats] List of mjd values of the bursts observed. Must NOT
                         be an astropy time quantity.
    
    
    RETURNS:
    
    TalliedEvents : [array of ints of shape equal to list of observing times]
    """
    
    
    TalliedEvents = np.zeros_like(ObservingTimes)

    #print(TalliedEvents)
    for i in range(len(BurstMJDs)): #loop over marcote bursts
        for j in range(len(ObservingTimes)): #loop over observation start times
            if (Time(BurstMJDs[i],format='mjd') > ObservingTimes[j]) & (Time(BurstMJDs[i],format='mjd') < ObservingTimes[j] + ObservingDurations[j]): #if burst falls into obs
                #print(True)
                TalliedEvents[j]+=1 #tally burst
    if np.sum(TalliedEvents) != len(BurstMJDs):
        print('warning! Not all bursts are tallied. Code is wrong.')
    else:
        print('bursts tallied.')
        
    return TalliedEvents
```

# Step 1: Recreate Effelsberg activity plots from Cruces+21

## Load data

```python
#set Cruces+21 -derived period
Cruces_p = 161 #[days]
```

```python
#import Cruces+21 data from table
Cruces_data = np.loadtxt('Cruces_Bursts.csv',dtype='str',skiprows=1,delimiter=',') #import all data
Cruces_mjds = np.array(Cruces_data[:,1],dtype=float) #select the mjd values

#import Hardy+17 data from table
Hardy_data = np.loadtxt('Hardy_Bursts.csv',dtype='str',skiprows=2,delimiter=',') #import all data
Hardy_mjds = np.array(Hardy_data[:,0]) #select the mjd values
Hardy_mjds = np.array([i.split('(')[0] for i in Hardy_mjds],dtype=float) #for now, get rid of the rounded digits (may need to make precision better later)

#import Houben+19 data from table
Houben_data = np.loadtxt('Houben_Bursts.csv',dtype='str',skiprows=1,delimiter=',') #import all data
Houben_mjds = np.array(Houben_data[:,1],dtype=float) #select the mjd values
```

```python
#combine all data
all_mjds = np.concatenate((Cruces_mjds,Hardy_mjds,Houben_mjds))
all_mjds.sort()
print(all_mjds)
```

## Manually create phase plot

```python
#get the phase data manually

t0 = np.min(all_mjds) #initial time of first observation

tarr = all_mjds-t0 #arrival times after first observation 

phase = (tarr/Cruces_p)-(tarr//Cruces_p) #arrival time as a function of phase according to Cruces+21 period
```

```python
#plot the phase data
bins = np.linspace(0,1,40)

fig = plt.figure(figsize=(8,6))
ax1=fig.add_subplot(111)
ax1.set_title('Manual analysis',fontsize=20)
ax1.set_xlabel('Phase',fontsize=20)
ax1.set_ylabel('Burst Count',fontsize=20)
ax1.tick_params(axis='x',labelsize=20)
ax1.tick_params(axis='y',labelsize=20)

ax1.hist(phase,bins=bins)
plt.tight_layout()
plt.show()
```

## Check by repeating with astropy

```python
#fold with astropy
all_ts = ts(time=Time(all_mjds,format='mjd')) #convert to astropy timeseries

#fold over Cruces+21 period
all_ts_folded = all_ts.fold(period=Cruces_p*u.day) 

#turn into phase
phase = (all_ts_folded.time.jd-np.min(all_ts_folded.time.jd))/Cruces_p

#plot
fig = plt.figure(figsize=(8,6))
ax1=fig.add_subplot(111)
ax1.set_title('Astropy fold',fontsize=20)
ax1.set_xlabel('Phase',fontsize=20)
ax1.set_ylabel('Burst Count',fontsize=20)
ax1.tick_params(axis='x',labelsize=20)
ax1.tick_params(axis='y',labelsize=20)

ax1.hist(phase,bins=bins)
plt.tight_layout()
plt.show()
```

Result: Discuss, but I think the results of both methods are the same, and also the same as Cruces+21 (Fig. 5 bottom), just phase shifted.

# Step 2: Recreate Cruces+21 Lomb-Scargle periodicity

```python
#import Cruces+21 data from table
Cruces_obs = np.loadtxt('Cruces_Obs.csv',dtype='str',skiprows=2,delimiter=',') #import all data
Cruces_Eff_obs = np.array([i for i in Cruces_obs if 'EFF' in i]) #extract only effelsberg observations
Cruces_Eff_times = Cruces_Eff_obs[:,1] #get the start utc of the observations
Cruces_Eff_times = Time(Cruces_Eff_times) #convert to astropy times
Cruces_Eff_events = np.array(Cruces_Eff_obs[:,-1],dtype=int) #get the number of events detected per observation
Cruces_Eff_durations = np.array(Cruces_Eff_obs[:,2],dtype=float)*u.s #get the duration of each observation in astropy s

#import Hardy+17 data from table
Hardy_obs = np.loadtxt('Hardy_Obs.csv',dtype='str',skiprows=2,delimiter=',') #import all data
Hardy_Eff_obs = Hardy_obs #all Hardy+17 observations were performed using the effelsberg
Hardy_Eff_times = Time(np.array(['{0} {1}'.format(i[0],i[1]) for i in Hardy_Eff_obs])) #convert to astropy times
print(Hardy_Eff_times)
Hardy_Eff_events = np.array(Hardy_obs[:,-1],dtype=int)
Hardy_Eff_Endtimes = Time(np.array(['{0} {1}'.format(i[0],i[2]) for i in Hardy_Eff_obs]))
Hardy_Eff_durations = (Hardy_Eff_Endtimes - Hardy_Eff_times).to(u.s) #get the duration of each observation in astropy s

#import Houben+19 data from table
Houben_obs = np.loadtxt('Houben_Obs.csv',dtype='str',skiprows=2,delimiter=',') #import all data
Houben_Eff_obs = np.array([i for i in Houben_obs if 'EFF' in i]) #extract only effelsberg observations
Houben_Eff_times = Time(np.array(['{0} {1}'.format(i[1],i[2]) for i in Houben_Eff_obs])) #convert to astropy times
Houben_Eff_events = np.array(Houben_Eff_obs[:,-1],dtype=int)
Houben_Eff_durations = np.array(Houben_Eff_obs[:,3],dtype=str) #get the duration of each observation
Houben_Eff_durations = [i.split(' ') for i in Houben_Eff_durations]
Houben_Eff_hours = np.array([i[0].split('h')[0] for i in Houben_Eff_durations],dtype=int)*u.hr
Houben_Eff_mins = np.array([i[1].split('m')[0] for i in Houben_Eff_durations],dtype=int)*u.min
Houben_Eff_durations = (Houben_Eff_hours+Houben_Eff_mins).to(u.s) #convert to astropy s
print(Houben_Eff_durations)
```

```python
#combine all data
all_Eff_times = Time(np.concatenate((Cruces_Eff_times,Hardy_Eff_times,Houben_Eff_times)))
all_Eff_events = np.concatenate((Cruces_Eff_events,Hardy_Eff_events,Houben_Eff_events))
all_Eff_durations = np.concatenate((Cruces_Eff_durations,Hardy_Eff_durations,Houben_Eff_durations))
```

```python
#get binary array of event occurences
eventcheck = np.array(all_Eff_events>0,dtype=int)
```

```python
#do lomb scargle periodicity search
#method from: https://jakevdp.github.io/blog/2015/06/13/lomb-scargle-in-python/

#choose periods to search
periods = np.logspace(0,3,500) #[days]

#get angular frequencies
ang_freqs = 2*np.pi/periods

#subtract mean as discussed in Cruces+21 and method
eventcheck_ms = eventcheck - np.mean(eventcheck)

#perform the lomb scargle periodicity search
power = ls(all_Eff_times.jd,eventcheck_ms,ang_freqs)

#plot results
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111)
ax.set_xlabel('Period [days]',fontsize=20)
ax.set_ylabel('Power',fontsize=20)
ax.set_ylim([0,2])
ax.tick_params(axis='x',labelsize=20)
ax.tick_params(axis='y',labelsize=20)
ax.set_xscale('log')
ax.yaxis.grid(True)

ax.plot(periods,power)
ax.axvline(Cruces_p,color='r',ls='--',label='Cruces+21 periodicity = {0} days'.format(Cruces_p))

plt.legend(fontsize=20)
plt.tight_layout()
plt.show()
```

```python
# cruces lomb scargle window function

#######################################
#get arrays to create window functions#
#######################################

#list of all days between first and last observing day
#(Note: this is a substitute for observing days)

window_days_Eff = np.arange(int(np.amin(all_Eff_times.jd)),
                                  int(np.amax(all_Eff_times.jd))+1)

#binary mask of days which had observations
#(Note: this is a substitute for events)

#initialise with zeros
window_obs_Eff = np.zeros_like(window_days_Eff,dtype=int)

#loop over observing days, set any day with an observation to one

#Arecibo data only
for i in range(len(all_Eff_times.jd)):
    
    #get day (before decimal point only)
    obs_day = int(all_Eff_times.jd[i])
    
    #get index of this day in list of all days
    idx = np.where(window_days_Eff==obs_day)
    
    #set this day to 1 in initialised mask
    window_obs_Eff[idx]=1
    
#subtract mean as discussed in Cruces+21 and method
window_obs_ms_Eff = window_obs_Eff - np.mean(window_obs_Eff)

##########################
#choose periods to search#
##########################

periods = np.logspace(0,3,500) #[days]

#from periods, get angular frequencies
ang_freqs = 2*np.pi/periods

#############################################
#perform the lomb scargle periodicity search#
#############################################

window_power_Eff = ls(window_days_Eff,window_obs_ms_Eff,ang_freqs)

#plot results
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111)
ax.set_xlabel('Period [days]',fontsize=20)
ax.set_ylabel('Power',fontsize=20)
#ax.set_ylim([0,2])
ax.tick_params(axis='x',labelsize=20)
ax.tick_params(axis='y',labelsize=20)
ax.set_xscale('log')
ax.yaxis.grid(True)

ax.plot(periods,window_power_Eff)
ax.axvline(Cruces_p,color='r',ls='--',label='Cruces+21 periodicity = {0} days'.format(Cruces_p))

plt.legend(fontsize=20)
plt.tight_layout()
plt.show()
```

```python
#overplot these on twin axes to compare

#plot results
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111)
ax.set_xlabel('Period [days]',fontsize=20)
ax.set_ylabel('LS Power',fontsize=20)
ax.set_ylim([0,2])
ax.tick_params(axis='x',labelsize=20)
ax.tick_params(axis='y',labelsize=20)
ax.set_xscale('log')
ax.yaxis.grid(True)

ax.plot(periods,power,label='Lomb Scargle')
ax.axvline(Cruces_p,color='r',ls='--',label='Cruces+21 periodicity = {0} days'.format(Cruces_p))
ax.legend(fontsize=20)

ax2 = ax.twinx()
ax2.set_ylabel('Window Power',fontsize=20)
ax2.tick_params(axis='x',labelsize=20)
ax2.tick_params(axis='y',labelsize=20)
ax2.set_xscale('log')
ax2.yaxis.grid(True)

ax2.plot(periods,window_power_Eff,color='C1',alpha=0.5,label='Window function')


plt.legend(fontsize=20)
plt.tight_layout()
plt.show()
```

Result: I think this is also the same as Cruces+21 (Fig 5 top).

## Lomb scargle periodicity plot

```python
#method from: https://jakevdp.github.io/blog/2015/06/13/lomb-scargle-in-python/

############################################
#get arrays to use in Lomb Scargle analysis#
############################################

#get binary array of event occurences
eventcheck_Arecibo = np.array(all_Arecibo_events>0,dtype=int)
eventcheck_Arecibo_Eff = np.array(all_Arecibo_Eff_events>0,dtype=int)

#subtract mean as discussed in Cruces+21 and method
eventcheck_ms_Arecibo = eventcheck_Arecibo - np.mean(eventcheck_Arecibo)
eventcheck_ms_Arecibo_Eff = eventcheck_Arecibo_Eff - np.mean(eventcheck_Arecibo_Eff)


##########################
#choose periods to search#
##########################

periods = np.logspace(0,3,500) #[days]

#from periods, get angular frequencies
ang_freqs = 2*np.pi/periods

#############################################
#perform the lomb scargle periodicity search#
#############################################

power_Arecibo = ls(all_Arecibo_times.jd,eventcheck_ms_Arecibo,ang_freqs)
power_Arecibo_Eff = ls(all_Arecibo_Eff_times.jd,eventcheck_ms_Arecibo_Eff,ang_freqs)

##############
#plot results#
##############

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111)
ax.set_title('Lomb Scargle Outputs',fontsize=20)
ax.set_xlabel('Period [days]',fontsize=20)
ax.set_ylabel('Power',fontsize=20)
ax.tick_params(axis='x',labelsize=20)
ax.tick_params(axis='y',labelsize=20)
ax.set_xscale('log')
ax.yaxis.grid(True)

ax.plot(periods,power_Arecibo,label='Arecibo observations')
ax.plot(periods,power_Arecibo_Eff,alpha=0.5,label='Arecibo+Eff observations')

ax.axvline(Cruces_p,color='r',ls='--',label='Cruces+21 periodicity = {0} days'.format(Cruces_p))

plt.legend(fontsize=20)
plt.tight_layout()
plt.show()
```

# Lomb Scargle Window Functions

```python
#######################################
#get arrays to create window functions#
#######################################

#list of all days between first and last observing day
#(Note: this is a substitute for observing days)

window_days_Arecibo = np.arange(int(np.amin(all_Arecibo_times.jd)),
                                  int(np.amax(all_Arecibo_times.jd))+1)
window_days_Arecibo_Eff = np.arange(int(np.amin(all_Arecibo_Eff_times.jd)),
                                  int(np.amax(all_Arecibo_Eff_times.jd))+1)

#binary mask of days which had observations
#(Note: this is a substitute for events)

#initialise with zeros
window_obs_Arecibo = np.zeros_like(window_days_Arecibo,dtype=int)
window_obs_Arecibo_Eff = np.zeros_like(window_days_Arecibo_Eff,dtype=int)

#loop over observing days, set any day with an observation to one

#Arecibo data only
for i in range(len(all_Arecibo_times.jd)):
    
    #get day (before decimal point only)
    obs_day = int(all_Arecibo_times.jd[i])
    
    #get index of this day in list of all days
    idx = np.where(window_days_Arecibo==obs_day)
    
    #set this day to 1 in initialised mask
    window_obs_Arecibo[idx]=1

#Arecibo + Effelsberg data
for i in range(len(all_Arecibo_Eff_times.jd)):
    
    #get day (before decimal point only)
    obs_day = int(all_Arecibo_Eff_times.jd[i])
    
    #get index of this day in list of all days
    idx = np.where(window_days_Arecibo_Eff==obs_day)
    
    #set this day to 1 in initialised mask
    window_obs_Arecibo_Eff[idx]=1
    
#subtract mean as discussed in Cruces+21 and method
window_obs_ms_Arecibo = window_obs_Arecibo - np.mean(window_obs_Arecibo)
window_obs_ms_Arecibo_Eff = window_obs_Arecibo_Eff - np.mean(window_obs_Arecibo_Eff)

##########################
#choose periods to search#
##########################

periods = np.logspace(0,3,500) #[days]

#from periods, get angular frequencies
ang_freqs = 2*np.pi/periods

#############################################
#perform the lomb scargle periodicity search#
#############################################

window_power_Arecibo = ls(window_days_Arecibo,window_obs_ms_Arecibo,ang_freqs)
window_power_Arecibo_Eff = ls(window_days_Arecibo_Eff,window_obs_ms_Arecibo_Eff,ang_freqs)

##############
#plot results#
##############

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111)
ax.set_title('Lomb Scargle Window Functions',fontsize=20)
ax.set_xlabel('Period [days]',fontsize=20)
ax.set_ylabel('Power',fontsize=20)
ax.tick_params(axis='x',labelsize=20)
ax.tick_params(axis='y',labelsize=20)
ax.set_xscale('log')
ax.yaxis.grid(True)

ax.plot(periods,window_power_Arecibo,label='Arecibo LS Window Function')
ax.plot(periods,window_power_Arecibo_Eff,alpha=0.5,label='Arecibo+Eff LS Window Function')

ax.axvline(Cruces_p,color='r',ls='--',label='Cruces+21 periodicity = {0} days'.format(Cruces_p))

plt.legend(fontsize=20)
plt.tight_layout()
plt.show()
```
